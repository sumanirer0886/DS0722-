```python
import pandas as pd

# Read the Excel file into a DataFrame
df1 = pd.read_excel('ProjectCSVPhaseI.csv')
# States  population  Litercay Rate   Area in this format
# Print the DataFrame
pop=df1['population']
LRate=df1['Litercay Rate']
Ar=df1['Area']
print(df)
```

```python
#2.1

import matplotlib.pyplot as plt

# Sample data
states = state_name
literacy_rate = LRate
crime_rate = TC

# Create the figure and the axes
fig, ax1 = plt.subplots()

# Plot the literacy rate on the first y-axis
ax1.plot(states, literacy_rate, 'b-', label='Literacy Rate')
ax1.set_xlabel('State')
ax1.set_ylabel('Literacy Rate (%)', color='b')
ax1.tick_params('y', colors='b')

# Create the second y-axis
ax2 = ax1.twinx()

# Plot the crime rate on the second y-axis
ax2.plot(states, crime_rate, 'r-', label='Crime Rate')
ax2.set_ylabel('Crime Rate', color='r')
ax2.tick_params('y', colors='r')

# Display the legend
lines = ax1.get_lines() + ax2.get_lines()
ax1.legend(lines, [line.get_label() for line in lines], loc='upper left')

# Set the title and display the graph
plt.title('State Literacy Rate vs Crime Rate')
plt.show()
```

```python
#2.3

YwiseCrime=[]
for year in range(2001, 2013):
    
    column1_value = year  # Replace with the actual value from the first column
    column2_value = 'TOTAL'  # Replace with the actual value from the second column

    selected_data = df.loc[(df['YEAR'] == column1_value) & (df['DISTRICT'] == column2_value), 'TOTAL IPC CRIMES']
    selected_data = sum(selected_data)
    YwiseCrime.append(selected_data)

import matplotlib.pyplot as plt

# Sample data
years = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012]
crime_rates = YwiseCrime

# Create the graph
plt.plot(years, crime_rates, marker='o')
plt.xlabel('Year')
plt.ylabel('Crime Rate')
plt.title('Year vs Crime Rate')

# Display the graph
plt.show()
```
