```python
#3.1	Insert records from 42_District_wise_crimes_committed_against_women_2001_2012.csv into a table

import pandas as pd
import sqlite3

# Read the crime data file into a DataFrame
df = pd.read_csv('42_District_wise_crimes_committed_against_women_2001_2012.csv')

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Store the DataFrame in the database as a table
df.to_sql('crime_data', conn, index=False)


#3.2 Write SQL query to find the highest number of rapes & Kidnappings that happened in which state, District, and year



# Execute the SQL query to find the highest number of rapes and kidnappings
query = """
    SELECT "STATE/UT",  DISTRICT, Year, MAX(Rape) AS MaxRapes, MAX("Kidnapping and Abduction") AS MaxKidnappings
    FROM "crime_data"
    GROUP BY "STATE/UT",  DISTRICT, Year
    ORDER BY MaxRapes DESC, MaxKidnappings DESC
    LIMIT 1
"""

# Execute the query and fetch the result
result = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Display the result
print(result)
```

```python
#3.4 Insert records from 02_District_wise_crimes_committed_against_ST_2001_2012.csv into a new table

import pandas as pd
import sqlite3

# Read the crime data file into a DataFrame
df = pd.read_csv('02_District_wise_crimes_committed_against_ST_2001_2012.csv')

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Store the DataFrame in the database as a table
df.to_sql('crime_data', conn, index=False)

# 3.5 Write SQL query to find the highest number of dacoity/robbery in which district.
query = """
    SELECT DISTRICT, MAX(Robbery) AS MaxDacoityRobbery
    FROM crime_data
    WHERE DISTRICT <> 'TOTAL'
    GROUP BY DISTRICT
    ORDER BY MaxDacoityRobbery DESC
    LIMIT 1
"""

# Execute the query and fetch the result
result = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Display the result
print(result)
```

```python
# 3.6

import pandas as pd
import sqlite3

# Read the crime data file into a DataFrame
df = pd.read_csv('02_District_wise_crimes_committed_against_ST_2001_2012.csv')

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Store the DataFrame in the database as a table
df.to_sql('crime_data', conn, index=False)

# Execute the SQL query to find the districts with the lowest number of murders
query = """
    SELECT DISTRICT, MIN(Murder) AS MinMurders
    FROM crime_data
    GROUP BY DISTRICT
    HAVING MinMurders = (SELECT MIN(Murder) FROM crime_data)
"""

# Execute the query and fetch the results
result = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Display the results
print(result)
```

```python
# 3.7 Write SQL query to find the number of murders in ascending order in district and yearwise.

import pandas as pd
import sqlite3

# Read the crime data file into a DataFrame
df = pd.read_csv('02_District_wise_crimes_committed_against_ST_2001_2012.csv')

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Store the DataFrame in the database as a table
df.to_sql('crime_data', conn, index=False)

# Execute the SQL query to find the number of murders in ascending order in district and yearwise
query = """
    SELECT DISTRICT, Year, SUM(Murder) AS TotalMurders
    FROM crime_data
    GROUP BY DISTRICT, Year
    ORDER BY DISTRICT ASC, Year ASC
"""

# Execute the query and fetch the results
result = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Display the results
print(result)
```

```python
# 3.8.1	Insert records of STATE/UT, DISTRICT, YEAR, MURDER, ATTEMPT TO MURDER, and RAPE columns only from 01_District_wise_crimes_committed_IPC_2001_2012.csv into a new table

import sqlite3
import pandas as pd

# Read the CSV file into a data frame
data_frame = pd.read_csv('01_District_wise_crimes_committed_IPC_2001_2012.csv')

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Load the data frame into the database as a table
data_frame.to_sql('crime_data', conn, index=False)

# 3.8.2 Write SQL query to find which District in each state/ut has the highest number of murders yearwise. Your output should show STATE/UT, YEAR, DISTRICT, and MURDERS.

# Define the SQL query
sql_query = '''
SELECT
    "STATE/UT",
    "YEAR",
    "DISTRICT",
    "MURDERS"
FROM
    (
    SELECT
        "STATE/UT",
        "YEAR",
        "DISTRICT",
        "MURDERS",
        ROW_NUMBER() OVER (PARTITION BY "STATE/UT", "YEAR" ORDER BY "MURDERS" DESC) AS row_num
    FROM
        crime_data
    ) AS ranked
WHERE
    row_num = 1
'''

# Execute the SQL query and fetch the results
cursor = conn.execute(sql_query)
result = cursor.fetchall()

# Create a data frame from the query result
result_df = pd.DataFrame(result, columns=['STATE/UT', 'YEAR', 'DISTRICT', 'MURDERS'])

# Print the resulting data frame
print(result_df)
```

```python
# 3.8.3

import pandas as pd
import sqlite3

# Read the crime data file into a DataFrame
df = pd.read_csv('42_District_wise_crimes_committed_against_women_2001_2012.csv')

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Store the DataFrame in the database as a table
df.to_sql('crime_data', conn, index=False)


# Write SQL query to find the highest number of rapes & kidnappings that happened in which state, district, and year

# Execute the SQL query to find the highest number of rapes and kidnappings
query = """
    SELECT "STATE/UT", "DISTRICT", "Year", MAX("Rape") AS MaxRapes, MAX("Kidnapping and Abduction") AS MaxKidnappings
    FROM "crime_data"
    GROUP BY "STATE/UT", "DISTRICT", "Year"
    ORDER BY MaxRapes DESC, MaxKidnappings DESC
    LIMIT 1
"""

# Execute the query and fetch the result
result = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Display the result
print(result)

# Filter districts that appear in 3 or more years
filtered_result = result.groupby('DISTRICT').filter(lambda x: len(x) >= 3)

# Sort the filtered result in descending order by year
sorted_result = filtered_result.sort_values('Year', ascending=False)

# Print the sorted result
print(sorted_result[['STATE/UT', 'DISTRICT', 'MaxRapes', 'Year']])
```
